```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#read data from math class and portuguese class
data1 = pd.read_csv("student-mat.csv")
data2 = pd.read_csv("student-por.csv")

#concatenate it together
data= pd.concat([data1,data2],ignore_index = True)

#show large figure
plt.figure(figsize=(20,20))


#show correlation map
sns.heatmap(data.corr(),annot = True, fmt = ".2f",cbar = True)
plt.xticks(rotation = 90)
plt.yticks(rotation = 0)
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import scipy as scipy
import matplotlib.pyplot as plt


#read the data
data1 = pd.read_csv("student-mat.csv")
data2 = pd.read_csv("student-por.csv")


#for math class, calculate average
averageData1 = sum(data1.G3) / float(len(data1))
data1['ave_line'] = averageData1

data1['Average Math Final Grade'] = ['above_average' if i > averageData1 else 'below_average' for i in data1.G3]
#combine daily alcohol with weekly
data1['Dalc'] =  data1['Dalc'] + data1['Walc']

#show swarmplit
sns.swarmplot(x='Dalc',y = 'G3',hue = 'Average Math Final Grade',data = data1,palette = {'above_average': 'blue', 'below_average': 'red'})
plt.show()


#for portuguese class,calculate average
averageData2 = sum(data2.G3) / float(len(data2))
data2['ave_line'] = averageData2

data2['Average Portuguese Final Grade'] = ['above_average' if i > averageData2 else 'below_average' for i in data2.G3]
#combine datily alcohol with weekly
data2['Dalc'] =  data2['Dalc'] + data2['Walc']

#show swarmplot
sns.swarmplot(x='Dalc',y = 'G3',hue = 'Average Portuguese Final Grade',data = data2,palette = {'above_average': 'blue', 'below_average': 'red'})
plt.show()
```
